# Question 1

Plot the waveform for one sample data from each activity class. Are you able to see any difference/similarities between the activities? You can plot a subplot having 6 columns to show differences/similarities between the activities. Do you think the model will be able to classify the activities based on the data?

In this question, the csv files in the Combined/Train were read and stored in Numpy Array. We extracted data from a different time and offset so that we can remove the initial and ending noise in the data sample.


```python

folders = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
classes = {"WALKING":1,"WALKING_UPSTAIRS":2,"WALKING_DOWNSTAIRS":3,"SITTING":4,"STANDING":5,"LAYING":6}

combined_dir = "Combined"
dataset_dir = os.path.join(combined_dir, "Train")
X_train = []  # Sensor data
y_train = []  # Labels
time = 20
ignore_rows = 1000 #to ignore the inital noise of the sensor data

# Collecting all the data from the Combined Folders X_train
for folder in folders:
    folder_path = os.path.join(dataset_dir, folder)
    files = os.listdir(folder_path)
    
    # we took the 6th file for all activities
    file = files[5]
    df = pd.read_csv(os.path.join(dataset_dir, folder, file), sep=",", header=0)
    df = df[ignore_rows:ignore_rows + time * 50]

    X_train.append(df.values)
    y_train.append(classes[folder])
X_train = np.array(X_train)
y_train = np.array(y_train)

print("X_train shape:", X_train.shape) # 3D array
print("y_train shape:", y_train.shape) # 1D array

activities = ["WALKING",  "WALKING_DOWNSTAIRS","WALKING_UPSTAIRS","LAYING", "STANDING","SITTING", ]


fig, axes = plt.subplots(figsize=(15, 3), nrows=1, ncols=6) #subplot


for i, activity in enumerate(activities):
    # Found the index of first occurence of sample data and corresponds to each activity iterated.
    # Since y_train is a 1D array, we can use np.where to find the index of the first occurrence of the activity class.
    act_ind = np.where(y_train == (i + 1))[0][0]
    print(np.where(y_train == (i + 1)))
    sample_data = X_train[act_ind]

    # Plot the accelerometer data for each axis
    axes[i].plot(sample_data[:, 0], label='x', color='red')
    axes[i].plot(sample_data[:, 1], label='y', color='blue')
    axes[i].plot(sample_data[:, 2], label='z', color='green')

    # Set the title and labels
    axes[i].set_title(activity)
    axes[i].set_xlabel('Time Passed (ms)')
    axes[i].set_ylabel('Acceleration (g)')
    axes[i].legend()

# Adjust layout and display the plot

plt.show()
```

#### Observations

Following observations are general to all the sample data:

1. Laying: The acceleration in the x-axis is low, nearly 0. This may be due to the phone orientation being flat, which makes the x-axis horizontal. Comparing with dynamic sitautions, the x-axis acceleration is fairly stable. Here, the y-axis and z-axis acceleration show steady values due to gravity. The square roots of acceleration per axis squared is nearly 1g. This indicates that the device remains in a fixed orientation.

2. Standing: All of the three accelerations remain constant throught the time window. The z-axis is close to zero, y-axis is stable yet close to 0, however x-axis has an acceleration quite close to 1g. x-axis has a high value as the device shall be in a vertical position. The stability and lack of fluctuation clearly state it as a static activity.

3. Sitting: The graph of this activity performs sitting action from another position which causes a significant change in accelerations. As the person was going to sit, the device moves downwards thus an increase in the x-axis acceleration. However once the person has sat, x-axis returns back to its previous state. The y-axis and z-axis accelerations changes are not significant, they fluctuate in the way the person sits.

4. Walking: All three axes acceleration show strong and periodic oscillations. Firstly, the largest amplitude is of the x-axis, which can indicate repeated forward and backward motion during walking. The y-axis and z-axis oscillate as well but with smaller variables. The overall periodic pattern suggests the cyclic nature of walking. This pattern definitely distinguishes itself from the static activities previously mentioned.

5. Walking Downstairs: Here, the accelerations flutuate heavily and are less uniformly compared to normal walking. Additionally, the x-axis shows high -amplitude oscillations. The y-axis and z-axis fluctate with noticeably smaller variations.Such irregularity reflects the abrupt or tough nature of going down the stairs.

6. Walking Upstairs: Here, the oscillations are again strong with the x-axis peaking the highest to about 2g. The z-axis fluctuations are more pronouced showing the extra upward thrust during climbing up the stairs. Periodicity remains but with higher effort visible in the amplitude.



#### General Insights:

1. Static Activities(Laying, Sitting, Standing): Signals remain nearly constant and stable but different in axis offsets help us to differentiate each position

2. Dynamic Activities(Walking, Upstairs, Downstairs): Here Signals are oscillatory and irregular. Distinguishing them from static activites is fairly easier than distinguishing each of them among themselves as many of them have overlapping patterns.


#### Verdict:

An ML model can reliably distinguish between dynamic and static activities based on axis acceleration patterns. However, distinguishing activities withing themselves (static, dynamic) may require more features as accelerations do overlap in pattern but only differ subtly in irregularity and amplitudes.

```python

time = 10
offset = 0
folders = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
classes = {"WALKING":1,"WALKING_UPSTAIRS":2,"WALKING_DOWNSTAIRS":3,"SITTING":4,"STANDING":5,"LAYING":6}

combined_dir = os.path.join("Combined")


# Got the train dataset 
X_train=[]
y_train=[]
dataset_dir = os.path.join(combined_dir,"Train")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))

    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_train.append(df.values)
        y_train.append(classes[folder])

X_train = np.array(X_train)
y_train = np.array(y_train)

# Got the test dataset
X_test=[]
y_test=[]
dataset_dir = os.path.join(combined_dir,"Test")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))
    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50] # Focussed on the first 500 rows of the dataset
        X_test.append(df.values)
        y_test.append(classes[folder]) # Appended the corresponding dictionary key to the y_test

X_test = np.array(X_test)
y_test = np.array(y_test)

# entire dataset x and y

X = np.concatenate((X_train,X_test))
y = np.concatenate((y_train,y_test))


seed = 4
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=seed,stratify=y)

print("Training data shape: ",X_train.shape)

print("Testing data shape: ",X_test.shape)
```

```python
sampling_rate = 50
time = 10
#making a nested list to sort the data according to the activies
activities= [[] for i in range(6)]    

for sample in range(len(X_train)): #going through all samples
    index = y_train[sample]-1
    for lines in range(sampling_rate*time): #going through all rows
        acc = X_train[sample][lines]
        activities[index].append(acc[0]**2+acc[1]**2+acc[2]**2) #appending the total acceleration in activities list


print("dynamic")
print(f"Walking:            {np.average(activities[0])}")
print(f"Walking_upstairs:   {np.average(activities[1])}")
print(f"Walking_downstairs: {np.average(activities[2])}")
print(f"static:")
print(f"Sitting:  {np.average(activities[3])}") #the indexing is done in the same was as in the classes dicionary in the codeblock above
print(f"Standing: {np.average(activities[4])}")
print(f"Laying:   {np.average(activities[5])}")
print()




dynamic_average = np.average(np.array([np.average(activities[i]) for i in [0,1,2]]))
static_average = np.average(np.array([np.average(activities[i]) for i in [3,4,5]]))
print(f"\n threshold: {np.average(np.array([float(dynamic_average), float(static_average)]))}")
```

From the averages seen above, we do not require a ML Model for the classification of static vs dynamic activities. The threshold value calculated here as 1.10734018055803 is the average of static average and dynamic average, each being an average of their activities. We can simply use this threshold to say if it is lower than this threshold, it is a static activity while higher suggests a dynamic activity.
